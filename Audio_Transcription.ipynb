Whisper - Large version 2

```python
import whisper

model_4 = whisper.load_model("large-v2")
```

```python
result_4 = model_4.transcribe("agentinsights.wav")
```

```python
result_4['text']
```

```python
import json

# Extract the transcribed text
transcribed_text = result_4['text']

# Define the output JSON file name
output_file = 'transcript.json'

# Create a JSON object with the transcribed text
json_data = {
    "transcript": transcribed_text
}

# Save the JSON object to the JSON file
with open(output_file, 'w', encoding='utf-8') as json_file:
    json.dump(json_data, json_file, ensure_ascii=False, indent=4)

# Print a message to confirm the save
print(f'Transcription has been saved to {output_file}')
```
